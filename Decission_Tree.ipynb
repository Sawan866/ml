```python
#  Decission Tree
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import pydotplus
import matplotlib.pyplot as plt
import matplotlib.image as pltimg

```

```python
df = pd.read_csv('/content/shows.csv')
print(df)
```

```python
d = {'UK': 0, 'USA': 1, 'N': 2}
df[' Nationality']=df[' Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

print(df)
```

```python
# Seprating feature Column and Target Column
# x = feature, y= target
feature = ['Age',' Experience','Rank',' Nationality']

X=df[feature]
y=df['Go']

print(X)
print(y)
```

```python
# Create decision tree, save and show it as image
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X,y)
data = tree.export_graphviz(dtree, out_file=None, feature_names=feature)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('mydecisiontree.png')

img=pltimg.imread('mydecisiontree.png')
imgplot = plt.imshow(img)
plt.show()
```
